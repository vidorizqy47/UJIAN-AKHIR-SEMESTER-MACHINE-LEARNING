# import library

```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.colors import ListedColormap
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

# Import dataset

```python
dt = pd.read_excel("../input/data-penduduk/penduduk provinsi.xls")
```

# menampilkan 5 data teratas

```python
dt.head(5)
```

# menampilkan informasi dataset

```python
dt.info()
```

# proses penghilangan nilai NaN

```python
dt.dropna()
```

# menampilkan elemen array

```python
dt.shape
```

# hapus kolom yang tidak diperlukan

```python
da = dt.drop(['Provinsi'], axis = 1)
da.head()
```

# Ulangi proses penghilangan nilai NaN

```python
da.dropna(how='any') 
```

```python
X = pd.DataFrame(da.dropna(how='any'))
X.head()
```

# deskripsi gambaran dataset

```python
X.describe()
```

# proses melihat ekor data

```python
X.tail()
```

# proses menebak jumlah cluster

```python
scr=[]
for i in range (1, 33):
    score = KMeans(n_clusters=i).fit(X).score(X)
    print(score)
    scr.append(score)
```

# ambil nilai tengah cluster

```python
plt.plot(scr)
```

# ubah data frame menjadi array

```python
x_array = np.array(X)
print(x_array)
```

# scaling data array

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_array)
x_scaled
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
```

# Cari nilai pusat setiap clusternya

```python
print(kmeans.cluster_centers_)
```

# menampilkan hasil clustering

```python
print(kmeans.labels_)
    
# Menambahkan kolom "cluster" dalam dataframe
X["cluster"] = kmeans.labels_
X.head()
```

# memvisualisasikan pembagian cluster

```python
plt.hist(X["cluster"])
```

# visualisasi pengelompokan cluster

```python
plt.figure()
sns.pairplot(X, hue="cluster", dropna=True)
plt.show()
```
