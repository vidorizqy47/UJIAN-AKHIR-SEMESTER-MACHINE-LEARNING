```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  

# we need to install mlxtend on anaconda prompt by typing 'pip install mlxtend'
from mlxtend.frequent_patterns import apriori  
from mlxtend.frequent_patterns import association_rules

# add dataset
df = pd.read_csv('../input/orderdetails/order_details.csv')
```

```python
df.head(10)
```

```python
dt = df.groupby(['productID', 'unitPrice'])['quantity'].sum().unstack().fillna(0)
print(dt.head())
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

data_sets = dt.applymap(encode_units)
```

```python
frequent_itemsets = apriori(data_sets,min_support=0.03,  use_colnames=True)
print(frequent_itemsets)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```
