# import library

```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.colors import ListedColormap
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

# ambil data set

```python
dt = pd.read_excel("../input/uasml2a/uasdata1.xlsx", engine='openpyxl')
```

# menampilkan 5 data teratas

```python
dt.head(5)
```

# menampilkan informasi data set

```python
dt.info()
```

# menghapus variabel NaN (tidak mempunyai nilai)

```python
dt.dropna()
```

# menampilkan banyaknya elemen 

```python
dt.shape
```

# Hapus kolom kolom yang tidak penting

```python
da = dt.drop(['Provinsi'], axis = 1)
da.head()
```

# recheck data NaN

```python
da.dropna(how='any') 
```

```python
X = pd.DataFrame(da.dropna(how='any'))
X.head()
```

# Menggambarkan deskripsi data set

```python
X.describe()
```

# untuk melihat ekor data gunakan fungsi tail()

```python
X.tail()
```

# proses menebak jumlah cluster

```python
scr=[]
for i in range (1, 34):
    score = KMeans(n_clusters=i).fit(X).score(X)
    print(score)
    scr.append(score)
```

# menghambil nilai yang ada ditengah-tengah lekukan cluster

```python
plt.plot(scr)
```

# mengubah variabel menjadi array

```python
x_array = np.array(X)
print(x_array)
```

# proses scaling data array

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_array)
x_scaled
```

# # tentukan jumlah clusternya

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
```

# Cari nilai pusat tiap cluster

```python
print(kmeans.cluster_centers_)
```

# Tampilkan hasil clustering

```python
print(kmeans.labels_)
    
# Menambahkan kolom "cluster" dalam dataframe
X["cluster"] = kmeans.labels_
X.head()
```

# visualisasi pembagian cluster

```python
plt.hist(X["cluster"])
```

# visualisasi pengelompokan cluster

```python
plt.figure()
sns.pairplot(X, hue="cluster", dropna=True)
plt.show()
```
